### Exercicio 11
Faça um programa que leia a largura e a altura de uma parede em metros, calcule a sua área e a quantidade de tinta necessária para pintá-la, sabendo que cada litro de tinta pinta uma área de 2 metros quadrados.

```python
largura = float(input("Qual a largura da parede: "))
altura = float(input("Qual a altura da parede: "))
area = largura*altura
l_tinta = 2 # Um litro de tinta pinta x metros quadrados
qtd_tinta = area/l_tinta
print("A sua parede de dimensão: {:.2f} m x {:.2f} m tem uma área de {:.2f}m².".format(largura,altura,area))
print("Para essa área: {:.2f}m², você precisará de {:.1f} latas de tinta.".format(area, qtd_tinta))
```

### Exercicio 32
Faça um programa que leia um ano qualquer e mostre se ele é bissexto.

```python
#ano=int(input("Informe o ano que deseja saber: "))
bi = 0
ano = 1582
for ano in range(2023):
    if ano % 400 == 0 or (ano % 4 == 0 and ano % 100 != 0):
        print("O ano {} é um ano Bissexto!".format(ano))
        bi = bi+1 
    else:
        print("O ano {} não é um ano Bissexto!".format(ano))
print("Já tivemos \033[0;34m{}\033[m anos bissextos!".format(bi))
```
